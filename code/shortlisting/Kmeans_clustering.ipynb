```python
import random
import numpy as np
import pylab
import codecs
data = np.loadtxt("entity2vec.bern")
entity2id = {}
id2entity = {}
f = codecs.open("entity2id.txt",encoding="latin")
entities = f.read().split("\n")
for x in entities:
    #print(x)
    entity = x.split('\t')[0]
    num = (int)(x.split('\t')[1])
    entity2id[entity] = num
    id2entity[num] = entity
```

```python
def K_means(data,centroids,label):
    
    #Assignment Step
    new_label = np.zeros(len(data))
    for i,example in enumerate(data):
        lis = [np.linalg.norm(example-x) for x in centroids]
        new_label[i] = np.argmin(lis)
    
    #Update Step
    for i,point in enumerate(centroids):
            lis = np.where(new_label==i)[0]
            centroids[i] = np.mean(data[lis],axis=0)
    return centroids,new_label
```

```python
#initialize centroids
K = 20
T=1
centroids = np.zeros((K,data.shape[1]))
label = np.zeros(len(data))
new_label = np.zeros(len(data))
for i in range(len(data)):
    label[i] = random.randint(0,K-1)
    new_label[i] = random.randint(0,K-1)
num = random.sample(range(len(data)), K)
for i,j in enumerate(num):
    centroids[i] = data[j]

while np.array_equal(new_label,label)==False and T<=40:
    label = new_label
    centroids,new_label = K_means(data,centroids,label)
    T =T+1
    print(T)
```
